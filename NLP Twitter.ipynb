```python
from sklearn.feature_extraction.text import CountVectorizer
d=CountVectorizer()
d
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,roc_auc_score
X_train,X_test,y_train,y_test=train_test_split(df_train['text'],df_train['target'],random_state=0)

vec_train=d.fit_transform(X_train)
vec_test=d.transform(X_test)

model=MultinomialNB(alpha=1)
model.fit(vec_train,y_train)

y_pred=model.predict(vec_test)
print(cross_val_score(model,vec_train,y_train).mean())
print(roc_auc_score(y_test,y_pred))


```
